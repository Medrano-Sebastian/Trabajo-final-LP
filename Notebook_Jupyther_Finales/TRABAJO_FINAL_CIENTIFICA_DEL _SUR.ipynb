```python
import pandas as pd
import urllib.request, urllib.parse, urllib.error
import requests
import re
from bs4 import BeautifulSoup
# Realizamos la petición a la página web
url_or = 'https://repositorio.cientifica.edu.pe/'
url="https://repositorio.cientifica.edu.pe//handle/20.500.12805/699"
page = requests.get(url)
# Creamos un objeto BeautifulSoup con el contenido de la página
soup = BeautifulSoup(page.content, 'html.parser')
#print(soup.prettify())
z = soup.find_all("h4", class_='artifact-title')
#print(z)
links=[]
for i in z:
    #print(i.find("a",href=True)["href"])
    links.append(i.find("a",href=True)["href"])
data = {'Institución':'a' ,'Titulo': 'a','Autor(es)':'a','Grado': 'a','Asesores':'a','Resumen': 'a','Año': 'a'}
print(links)
lista1=[]
lista2=[]
listautor=[]
lista3=[]
lista4=[]
lista5=[]
lista6=[]
lista7=[]
for x in links:
    html = urllib.request.urlopen(url_or+x+"?show=full").read()
    sopas = BeautifulSoup(html, 'html.parser')
    #cosas=str(sopas.find_all("td", class_='word-break').text)
    #print(cosas)
    filas = sopas.find_all("tr")
    for fila in filas:
    # Encuentra todas las celdas de la fila
        celdas = fila.find_all("td", class_='label-cell')
        or_celd=fila.find_all("td", class_='word-break')
        # Itera sobre cada celda de la fila
        for celda in celdas:
        # Imprime el contenido de la celda
            #print(celda.text)
            for cel in or_celd:
                #print('-------------------------------')
                if celda.text=="dc.publisher":
                    lista1.append(cel.text)
                if celda.text=="dc.title":
                    lista2.append(cel.text)
                if celda.text=="dc.contributor.author":
                    #print(cel.text)
                    listautor.append(cel.text)
                if celda.text=="dc.type.other":
                    lista4.append(cel.text)
                if celda.text=="dc.contributor.advisor":
                    lista5.append(cel.text)
                if celda.text=="dc.description.abstract":
                    lista6.append(cel.text)
                if celda.text=="dc.date.issued":
                    lista7.append(cel.text)
    #print("||||||||||||||||")
    #print(listautor)
    lemen=" ".join(listautor)
    #print(lemen)
    lista3.append(lemen)
    lemen= None
    listautor.clear()
data['Institución']=lista1        
data['Titulo']=lista2
data['Autor(es)']=lista3
data['Grado']=lista4
data['Asesores']=lista5
data['Resumen']=lista6
data['Año']=lista7
#print(data)        
df = pd.DataFrame(data, columns = ['Institución','Titulo','Autor(es)','Grado','Asesores','Resumen','Año'])
df.to_csv('Tesis Cientifica del sur.csv',encoding='utf-8',index=True)
#print(data['Autor(es)'])
#print(len(lista3))
pd.read_csv('Tesis Cientifica del sur.csv')   
```
