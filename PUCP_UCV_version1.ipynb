```python
#titulo
import re
import ssl
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
print(links_tesis)
titulo =[]
for key,val in links_tesis.items():
    html = urllib.request.urlopen(repo+val+"?show=full",context=ctx).read().decode("UTF-8")
    #print(html)
    linkeado = re.findall('<title>(.*)</', html)
    print(linkeado)
    print(len(linkeado))
    distinct(titulo,linkeado)
        #print(link)
len(titulo)
```

```python
#fechas
import re
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
print(links_tesis)
fechas =[]
for key,val in links_tesis.items():
    html = urllib.request.urlopen(repo+val,context=ctx).read()
    #print(html)
    linkeado = re.findall('<meta content="(.*)" name="citation_date">', html.decode())
    print(linkeado)
    print(len(linkeado))
    distinct(fechas,linkeado)
        #print(link)
len(fechas)
```

```python
#resumen
import re
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
print(links_tesis)
resumen =[]
for key,val in links_tesis.items():
    html = urllib.request.urlopen(repo+val,context=ctx).read().decode("utf-8")
    print(html)
    linkeado = re.findall('<meta name="DCTERMS.abstract" content="(.*)" xml:lang="es_ES"', html.decode())
    print(linkeado)
    print(len(linkeado))
    distinct(resumen,linkeado)
        #print(link)
len(resumen)
```
