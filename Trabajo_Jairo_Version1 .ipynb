```python
# Buscar valores de enlace dentro de la entrada de URL
import urllib.request, urllib.parse, urllib.error
import re
import ssl
import pandas as pd
from bs4 import BeautifulSoup
# Ignorar los errores del certificado SSL
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
url="https://tesis.pucp.edu.pe/repositorio/handle/20.500.12404/15964"
linkss="https://tesis.pucp.edu.pe/repositorio/"
i=url
b=0
i=i+"/recent-submissions?offset="
nombres=[]
b=0
for z in range(2):
    b=str(b)
    pag=i+b
    html = urllib.request.urlopen(pag, context=ctx).read()
    links = re.findall(b'href=.*/(handle/.*/\d*)".*Z3988', html)
    for link in links:
        o=linkss+link.decode()
        print(o)
        html = urllib.request.urlopen(o, context=ctx).read()
        html = BeautifulSoup(html, 'html.parser')
        html=str(html)
        links = re.findall(r'title>(.*)</title', html)
        for link in links:
            u=link
            print(u)
            nombres.append(u)
            

    b=int(b)
    b=b+20
df = pd.DataFrame(nombres, columns = ['TESIS'])


df.to_csv('TESISnombres1.csv',sep=";",encoding="latin-1")

```
